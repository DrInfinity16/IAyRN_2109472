```python
#Cargar librerias
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
import sklearn.linear_model # contiene diferentes modelos de ML
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#Cargar datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
# Preprocesamiento de los datos
x_train, x_test, y_train, y_test = train_test_split(horas_estudio.reshape(-1, 1), calificaciones, test_size=0.2, random_state=0);
```

```python
# Crear el modelo y entrenarlo
modelo = LinearRegression()
modelo.fit(x_train, y_train)
```

```python
# Hacer predicciones
predicciones = modelo.predict(x_test)

# Error cuadrático medio
error = mean_squared_error(y_test, predicciones)
print("Error cuadrático medio:", error)

# Graficar
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, predicciones, color='blue', linewidth=2)
plt.title('Regresión Lineal: Horas de estudio vs. calificación')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificación')
plt.grid(True)
plt.show()
```
